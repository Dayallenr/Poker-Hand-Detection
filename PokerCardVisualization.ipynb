```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
model = YOLO('/content/PokerDetectionBest.pt')
```

```python
results = model.predict('/content/IMG_5245.HEIC')
plotted = results[0].plot()
plt.imshow(cv2.cvtColor(plotted, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
results[0].summary()
```
